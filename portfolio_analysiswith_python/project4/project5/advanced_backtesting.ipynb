Buy when: SMA_1 > SMA_2 

Sell when: SMA_1 > SMA_2 

```python
# importing libnaries
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0].Symbol.tolist()
```

```python
price_df = yf.download(tickers, start = '2015-01-01', auto_adjust=False)
```

```python
price_df
```

```python
# Creating a finction to slice the dataframe on the Symbols
price_df[price_df.columns[price_df.columns.get_level_values(1) == 'AAPL']]
```

```python
# Creating the slicer function
def slice_df(symbol):
    sliced = price_df.copy()
    sliced = sliced[sliced.columns[sliced.columns.get_level_values(1)== symbol]]
    sliced.columns = sliced.columns.droplevel(1)
    sliced.loc[:, 'price'] = sliced.Open.shift(-1)
    return sliced
```

```python
# Calling the function
test = slice_df('AAPL')
```

```python
# Creating the movaing average function
def ma_calc(df, n, m): #n short term window, m long term window
    df['sma_1'] = df.Close.rolling(n).mean()
    df['sma_2'] = df.Close.rolling(m).mean()
```

```python
ma_calc(test, 50, 100)
test.dropna(inplace=True)
test
```

### Backtest: 
We are going to check the SMA_1 to SMA_2 value and buy when the next day if the SMA_1 value is greater.

Similarly, if SMA_2 value is greater we are goung to sell the position. 

```python
# We are going to backtest all the sub df, and their componenets
# We will create a function to backtest the strategy
def backtest(df, n, m):
    ma_calc(df, n, m)
    in_position = False
# Since there will be multuple trades, we will create a list to store the profits
    profits = []

    for index, row in df.iterrows():
        if not in_position:
            if row.sma_1 > row.sma_2:
                buy_price = row.price
                in_position = True
        if in_position:
            if row.sma_1 < row.sma_2:
                profit = (row.price - buy_price)/buy_price
                profits.append(profit)
                in_position = False

## Once all the trades are dine we will accumuklate the profits
    gain = (pd.Series(profits)+1).prod()
    return gain
```

```python
backtest(test,50,100)
```

This suggests for every dollar we would have invested we would have made a profit of $2.

Now, we are going to apply the function on all the symbols and store them under results.

```python
results = []

for sym in tickers:
    subdf = slice_df(sym)
    print('Profit realised for ' + sym)
    print(backtest(subdf, 50, 100))
    results.append(backtest(subdf, 50, 100))
```

```python
# Creating a dataframe to store the results
results_df = pd.DataFrame(results, index=tickers, columns=['Profit'])
results_df
```

```python
# largest profit realised
results_df.nlargest(50, 'Profit')
```

```python
# lets create the vector function

def vectorized(df, n, m):
    ma_calc(df, n, m)
    first_buy = pd.Series(df.index == (df.sma_1 > df.sma_2).idxmax(), index=df.index)
    real_signal = first_buy | (df.sma_1> df.sma_2).diff()
    trades = df[real_signal]

    if len(trades) % 2 != 0:
        mtm = df.tail(1).copy()
        mtm.price = mtm.close
        trades = pd.concat([trades, mtm])
    profits = trades.price.diff()[1::2] / trades.price[0::2].values
    gain = (profits + 1).prod()
    return gain
```
