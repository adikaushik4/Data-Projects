### Combining Dataframes
#### Concat, Merge & Join

```python
# Concat is used to join Datafranmes along 'rows'
# Concat does not merge indicex

sam_df1 = pd.DataFrame([1,2,3], columns=['test'])
sam_df2 = pd.DataFrame([3,4])
```

```python
pd.concat([ sam_df2, sam_df1], axis=1, ignore_index=True)
```

```python
# Merge is used to join Dataframes based on 'common columns'
sam_df1 = pd.DataFrame({'ID':[1,2,3], 'Name': ['A', 'B', 'C']})
sam_df3 = pd.DataFrame({'ID': [2, 3, 4], 'Score': [90, 85, 88]})
# Inner
df_merge = pd.merge(sam_df1, sam_df3, on='ID', how='inner')
print(df_merge)
```

### Plotting the returns of the S&P 500 since 2010

```python
df = yf.download('^GSPC',start='2009-12-31', auto_adjust=False)
```

```python
df.head()
```

```python
df.shift(1)
```

```python
# Calculating returns
df['returns'] = df.Close.pct_change() + 1
df.dropna(inplace=True)
df
```

```python
# Calculating the returns yearly
ret_yrl_sp= (df.resample('YE').apply(np.prod) - 1)['returns']
```

```python
ret_yrl_sp.plot(kind='bar')
```
